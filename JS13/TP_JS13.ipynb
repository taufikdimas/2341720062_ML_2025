# Tugas Praktikum
Gunakan JST untuk klasifikasi angka tulisan tangan (MNIST).

Langkah:

- Load dataset MNIST dari Keras.
- Bangun model dengan 2 hidden layer.
- Latih model dan evaluasi akurasi.

## Model dasar: 2 Hidden Layers (128 dan 64 neuron)

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import time

# 1. Load dataset MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalisasi data (0-255 → 0-1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encoding label
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 2. Bangun model JST
model = Sequential([
    Flatten(input_shape=(28, 28)),  # Ubah gambar 28x28 menjadi vektor
    Dense(128, activation='relu'),   # Hidden layer 1
    Dense(64, activation='relu'),    # Hidden layer 2
    Dense(10, activation='softmax')  # Output layer (10 kelas)
])

# 3. Kompilasi model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# 4. Latih model
print("Training Model Dasar (128, 64 neuron)...")
start_time = time.time()
history = model.fit(X_train, y_train,
                    epochs=10,
                    batch_size=128,
                    validation_split=0.1,
                    verbose=1)
training_time = time.time() - start_time

# 5. Evaluasi model
loss, acc = model.evaluate(X_test, y_test)
print(f"\nModel Dasar (128, 64 neuron)")
print(f"Akurasi pada data uji: {acc:.4f}")
print(f"Waktu pelatihan: {training_time:.2f} detik")
```

## Ubah jumlah neuron di hidden layer (misal: 256 dan 128)

> Model dengan 2 hidden layer: 256 & 128 (ReLU)

```python
# Soal 1: Model MLP dengan 2 hidden layer (256, 128)

model_1 = Sequential([
    # Ubah gambar 28x28 menjadi vektor
    Flatten(input_shape=(28, 28), name="flatten"),

    # Hidden layer 1: 256 neuron
    Dense(256, activation='relu', name="hidden_1"),

    # Hidden layer 2: 128 neuron
    Dense(128, activation='relu', name="hidden_2"),

    # Output layer: 10 neuron (softmax untuk klasifikasi 10 kelas)
    Dense(10, activation='softmax', name="output")
])

# 3. Kompilasi model
model_1.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# 4. Latih model
start_time = time.time()
history_1 = model_1.fit(
    X_train, y_train,
    epochs=10,
    batch_size=128,
    validation_split=0.1,
    verbose=1
)
train_time_1 = time.time() - start_time

# 5. Evaluasi model
loss_1, acc_1 = model_1.evaluate(X_test, y_test, verbose=0)
print(f"\n[Soal 1] Akurasi pada data uji: {acc_1:.4f}")
print(f"[Soal 1] Waktu pelatihan: {train_time_1:.2f} detik")
```

## Tambahkan satu hidden layer lagi

> Model dengan 3 hidden layer (256, 128, 64)

```python
# Soal 2: Tambahkan 1 hidden layer lagi → 3 hidden layer total

model_2 = Sequential([
    Flatten(input_shape=(28, 28), name="flatten"),

    Dense(256, activation='relu', name="hidden_1"),
    Dense(128, activation='relu', name="hidden_2"),
    Dense(64,  activation='relu', name="hidden_3"),  # hidden tambahan

    Dense(10, activation='softmax', name="output")
])

model_2.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

start_time = time.time()
history_2 = model_2.fit(
    X_train, y_train,
    epochs=10,
    batch_size=128,
    validation_split=0.1,
    verbose=1
)
train_time_2 = time.time() - start_time

loss_2, acc_2 = model_2.evaluate(X_test, y_test, verbose=0)
print(f"\n[Soal 2] Akurasi pada data uji: {acc_2:.4f}")
print(f"[Soal 2] Waktu pelatihan: {train_time_2:.2f} detik")
```

## Bandingkan akurasi dan waktu pelatihan

```python
def build_and_train(hidden_layers, label,):
    model = Sequential()
    model.add(Flatten(input_shape=(28, 28), name="flatten"))

    # Tambah hidden layer sesuai tuple hidden_layers
    for i, units in enumerate(hidden_layers, start=1):
        model.add(Dense(units, activation='relu', name=f"hidden_{i}"))

    model.add(Dense(10, activation='softmax', name="output"))

    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    print(f"\n=== Training {label} dengan hidden layer {hidden_layers} ===")
    start = time.time()
    history = model.fit(
        X_train, y_train,
        epochs=10,
        batch_size=128,
        validation_split=0.1,
        verbose=1
    )
    duration = time.time() - start

    loss, acc = model.evaluate(X_test, y_test, verbose=0)
    print(f"[{label}] Akurasi data uji: {acc:.4f}")
    print(f"[{label}] Waktu pelatihan: {duration:.2f} detik")

    return acc, duration

# Model A: lebih kecil
acc_A, time_A = build_and_train((128, 64), label="Model A (kecil)")

# Model B: lebih besar / lebih dalam
acc_B, time_B = build_and_train((256, 128, 64), label="Model B (besar)")

print("\n=== Ringkasan Perbandingan Soal 3 ===")
print(f"Model A (128, 64)      -> Akurasi: {acc_A:.4f}, Waktu: {time_A:.2f} detik")
print(f"Model B (256, 128, 64) -> Akurasi: {acc_B:.4f}, Waktu: {time_B:.2f} detik")
```

## Eksperimen dengan fungsi aktivasi Sigmoid vs ReLU

```python
def build_and_train_activation(hidden_layers, activation, label,):
    model = Sequential()
    model.add(Flatten(input_shape=(28, 28), name="flatten"))

    for i, units in enumerate(hidden_layers, start=1):
        model.add(Dense(units, activation=activation, name=f"{activation}_hidden_{i}"))

    model.add(Dense(10, activation='softmax', name="output"))

    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    print(f"\n=== Training {label} dengan aktivasi {activation} ===")
    start = time.time()
    history = model.fit(
        X_train, y_train,
        epochs=10,
        batch_size=128,
        validation_split=0.1,
        verbose=1
    )
    duration = time.time() - start

    loss, acc = model.evaluate(X_test, y_test, verbose=0)
    print(f"[{label}] Akurasi data uji: {acc:.4f}")
    print(f"[{label}] Waktu pelatihan: {duration:.2f} detik")

    return acc, duration

hidden_cfg = (128, 64)

# Model dengan ReLU
acc_relu, time_relu = build_and_train_activation(
    hidden_layers=hidden_cfg,
    activation='relu',
    label='Model ReLU'
)

# Model dengan Sigmoid
acc_sigmoid, time_sigmoid = build_and_train_activation(
    hidden_layers=hidden_cfg,
    activation='sigmoid',
    label='Model Sigmoid'
)

print("\n=== Ringkasan Perbandingan Soal 4 (ReLU vs Sigmoid) ===")
print(f"ReLU    -> Akurasi: {acc_relu:.4f}, Waktu: {time_relu:.2f} detik")
print(f"Sigmoid -> Akurasi: {acc_sigmoid:.4f}, Waktu: {time_sigmoid:.2f} detik")
```
